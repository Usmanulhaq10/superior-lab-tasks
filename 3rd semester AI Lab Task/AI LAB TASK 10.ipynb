```python
import os
import pandas as pd

folder_path = r'C:\Users\Chusm\OneDrive\Desktop\BSAIM-3A Usman 035\Artificial Intelligence - Lab\Task 9-14\Task 9'

csv_files_list = []
for file in os.listdir(folder_path):
    if file.endswith('.csv'):
        csv_files_list.append(file)

print("Files in the folder:", csv_files_list)

file_to_open = os.path.join(folder_path, 'netflix_titles.csv')
netflix_df = pd.read_csv(file_to_open)

print("\nFirst 5 rows:")
print(netflix_df.head())

print("\nData info:")
print(netflix_df.info())

print("\nNumber of rows:", netflix_df.shape[0])
print("Number of columns:", netflix_df.shape[1])

print("\nColumn names:")
print(netflix_df.columns)

print("\nSummary stats:")
print(netflix_df.describe(include='all'))

print("\nMissing values:")
print(netflix_df.isnull().sum())

print("\nUnique values per column:")
for column in netflix_df.columns:
    print(column, "has", netflix_df[column].nunique(), "unique values")

print("\nTypes count:")
print(netflix_df['type'].value_counts())

print("\nRatings count:")
print(netflix_df['rating'].value_counts())

print("\nMovies per year:")
print(netflix_df['release_year'].value_counts().sort_index())

netflix_df = netflix_df.drop_duplicates()

netflix_df['country'] = netflix_df['country'].fillna('Unknown')
netflix_df['director'] = netflix_df['director'].fillna('Unknown')
netflix_df['cast'] = netflix_df['cast'].fillna('Unknown')
netflix_df['date_added'] = netflix_df['date_added'].fillna(method='ffill')
netflix_df['rating'] = netflix_df['rating'].fillna('Not Rated')
netflix_df['duration'] = netflix_df['duration'].fillna('0 min')

netflix_df['release_year'] = netflix_df['release_year'].astype(int)

duration_numbers = netflix_df['duration'].str.extract('(\d+)')
duration_numbers = duration_numbers.astype(float)
duration_numbers = duration_numbers.fillna(0)
duration_numbers = duration_numbers.astype(int)
netflix_df['duration_minutes'] = duration_numbers

print("\nData types after changes:")
print(netflix_df.dtypes)

print("\nRelease year and duration preview:")
print(netflix_df[['release_year', 'duration_minutes']].head())
```
